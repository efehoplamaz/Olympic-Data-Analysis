```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import statistics
```

```python
athlete_events_df = pd.read_csv('olympic_data/athlete_events.csv')
noc_regions_df = pd.read_csv('olympic_data/noc_regions.csv')


athlete_events_df.head()
```

```python
sport_groupby = athlete_events_df.groupby('Sport')

dict_sports_mean_per_year = {}

for name, group in sport_groupby:
    
    mean_dict = {}
    mean_list = []
    
    year_groupby = group.groupby('Year')
    
    for year, year_group in year_groupby:

        mean_dict[year] = year_group.Age.mean()
        mean_list.append(year_group.Age.mean())
    
    mean_dict['mean'] = statistics.mean(mean_list)
    
    if len(mean_list) > 1:
        mean_dict['stdev'] = statistics.stdev(mean_list)
    else:
        mean_dict['stdev'] = 0
    
    dict_sports_mean_per_year[name] = mean_dict
    
l = sorted(dict_sports_mean_per_year.items(), key=lambda x: x[1]["stdev"])
```

```python
plt.figure(figsize=(8, 6), dpi=80)

for sports, d in l[-3:]:
    x = []
    y = []
    for year, mean in d.items():
        
        if year == 'mean' or year == 'stdev':
            continue
        x.append(year)
        y.append(mean)

    plt.plot(x, y, '-o', label = sports)

plt.xlabel('Year')
plt.ylabel('Mean age')
plt.legend()
plt.show()
```

```python
### EN YALI OLMPYAT + EN YKSEK STANDART SAPMA

year_grouping = athlete_events_df.groupby('Year')

d_year_mean_std = {}

for year, group in year_grouping:
    year_dict = {}
    mean_olympic_age_by_year = group.Age.mean()
    std_olympic_age_by_year = group.Age.std()
    
    year_dict['mean'] = mean_olympic_age_by_year
    year_dict['std'] = std_olympic_age_by_year
    
    d_year_mean_std[year] = year_dict

l = sorted(d_year_mean_std.items(), key = lambda x: x[1]['mean'])

oldest_year_mean = l[-1][0]

df_oldest_mean = year_grouping.get_group(oldest_year_mean)

age_list = [elm for elm in df_oldest_mean.Age.to_list() if math.isnan(elm) == False]

y = [0] * len(age_list)

plt.figure(figsize=(12, 6), dpi=80)
plt.tick_params(left = False, labelleft = False)
plt.scatter(age_list, y)
plt.scatter(df_oldest_mean.Age.quantile(0.25), 0, color = 'r')
plt.scatter(df_oldest_mean.Age.quantile(0.5), 0, color = 'r')
plt.scatter(df_oldest_mean.Age.quantile(0.75), 0, color = 'r')
```

```python
### SPORLARA GRE BOX WHISKER PLOT

list_of_popular_sports = list(dict(athlete_events_df.Sport.value_counts()).keys())[:10]

for sports in list_of_popular_sports[:5]:
    plt.figure(figsize=(8, 6), dpi=80)
    sport_dataframe = sport_groupby.get_group(sports)
    female_list = [elm for elm in sport_dataframe[(sport_dataframe.Sex == 'F')].Age.to_list() if math.isnan(elm) == False]
    male_list = [elm for elm in sport_dataframe[(sport_dataframe.Sex == 'M')].Age.to_list() if math.isnan(elm) == False]
    plt.boxplot([male_list, female_list], vert = 0, notch=True, patch_artist=True)
    plt.yticks([1, 2], ["Male " + sports, "Female " + sports])
    plt.title(sports)
    plt.show()
```
